# Basic Data Manipulation (Pandas, Numpy)

## Loading and Inspecting Data

```python
import pandas as pd
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
df = pd.read_csv(url, names=names)
```

```python
df['class'].value_counts()
```

## Manipulating the data
Add a new column that is the ratio of petal length to petal width:

```python
df['petal_ratio'] = df['petal_length'] / df['petal_width']

df.columns
```

Sort the DataFrame by this new column:

```python
df.sort_values('petal_ratio', ascending=False, inplace=True)
df.head()
```

# Basic Data Visualization (matplotlib, seaborn)

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.hist(df['sepal_length'], bins=20)
plt.xlabel('Sepal Length')
plt.ylabel('Count')
plt.title('Histogram of Sepal Length')
plt.show()
```

Boxplot with seaborn

Boxplots can provide more detailed statistical information:

```python
sns.boxplot(x='class', y='sepal_length', data=df)
plt.title('Boxplot of Sepal Length by Class')
plt.show()
```

Scatterplot with seaborn

Scatterplots can help visualize the relationship between two variables:

```python
sns.scatterplot(x='sepal_length', y='sepal_width', hue='class', data=df)
plt.title('Scatterplot of Sepal Length vs Sepal Width')
plt.show()
```

Pairplot with seaborn

seaborn's pairplot function can create a grid of scatterplots and histograms to show pairwise relationships and distributions for multiple variables at once:

```python
sns.pairplot(df, hue='class')
plt.show()
```

The `pairplot` function plots pairwise relationships across the entire dataframe (for the numerical columns) and supports a color hue argument (for categorical columns). Here's what it does:

*   It creates a grid of Axes such that each numeric variable in data will by shared in the y-axis across a single row and in the x-axis across a single column.
*   The diagonal Axes are treated differently, drawing a plot to show the univariate distribution of the data for the variable in that column.
*   Off-diagonal axes show the scatter plots, depicting the relationship (or lack thereof) between variables.

In our iris example, the pairplot function is plotting pairwise relationships across all numerical columns (i.e., 'sepal\_length', 'sepal\_width', 'petal\_length', 'petal\_width').

The `hue='class'` argument is saying that the colors of the points should be based on the 'class' column (i.e., the species of the iris flowers).

This `pairplot` allows us to see at a glance how the measurements for each pair of features differ by species, which can give us insights into potential relationships or correlations between variables. For instance, we can see that petal length and petal width seem to have a clear positive correlation for all species, and that these two features can separate Iris-setosa from the other two species quite well.
